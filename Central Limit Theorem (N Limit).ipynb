```python
"""
Central Limit Theorem (CLT), which states that the distribution of the sum (or average) of a large number 
of independent, identically distributed random variables, regardless of their original distribution, 
tends to follow a normal (Gaussian) distribution.  The theorem is a key concept in probability theory 
because it implies that probabilistic and statistical methods that work for normal distributions can be applicable
to many problems involving other types of distributions. In this example, we generate binomial data for 
different values of n and plot both the binomial histograms and the corresponding normal distributions on the 
same graph. As n increases, you will observe that the binomial distribution starts to resemble a normal 
distribution, demonstrating the convergence towards normality as predicted by the Central Limit Theorem.
"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# Set parameters
p = 0.5  # Probability of success
max_n = 500  # Maximum value of n

# Generate binomial data for different values of n
n_values = [10, 50, 100, 500]

# Set up the plot
plt.figure(figsize=(10, 6))

# Plot histograms and normal distributions
for n in n_values:
    # Generate binomial data
    binomial_data = np.random.binomial(n, p, 1000)

# Plot binomial histogram
    plt.hist(binomial_data, bins=np.arange(0, n + 2) - 0.5, density=True, alpha=0.5, label=f'Binomial (n={n})')
    
    # Calculate parameters for the corresponding normal distribution
    mu = n * p
    sigma = np.sqrt(n * p * (1 - p))
    
    # Plot normal distribution
    x = np.linspace(0, n, 1000)
    y = norm.pdf(x, mu, sigma)
    plt.plot(x, y, label=f'Normal (mean={mu}, std={sigma})', linestyle='dashed')

# Add labels and legend
plt.title('Comparison of Binomial and Normal Distributions')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()

# Show the plot
plt.show()
```
