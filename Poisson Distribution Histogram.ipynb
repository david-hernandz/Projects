```python
import matplotlib.pyplot as plt
import numpy as np

lambda_parameter = 5
data = np.random.poisson(lambda_parameter, 1000)

plt.hist(data, bins=np.arange(0, max(data) + 1.5) - 0.5, density=True, alpha=0.75)

plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("Poisson Distribution Histogram")

plt.show()
```
